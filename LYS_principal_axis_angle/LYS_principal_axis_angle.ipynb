```python
import pytraj as pt
import numpy as np
import scipy as sp
from scipy.interpolate import CubicSpline
import pandas as pd
import collections
import matplotlib
from matplotlib import pyplot as plt
from ipywidgets import FloatProgress
from IPython.display import display
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib.font_manager import FontProperties
from scipy.spatial import Delaunay
```

```python
negDataL=pd.read_table('neg200mV_LYS.txt',header=None,delim_whitespace=True)
frames=np.arange(0,9919)
print frames.shape
negDataL.columns=['LYS41','LYS267','LYS493','LYS719','LYS945','LYS1171']
negDataL['Frame']=np.transpose(frames)
display(negDataL.head())
listID=['LYS41','LYS267','LYS493','LYS719','LYS945','LYS1171']
```

```python
plt.rcParams['axes.facecolor']='white'
import matplotlib
matplotlib.rc('axes',edgecolor='k')
from matplotlib.pyplot import figure

listID=['LYS41','LYS267','LYS493','LYS719','LYS945','LYS1171']
print ('LYS orientation time frames with -200mV')
for iid in listID:
    figure(num=None, figsize=(12, 6), dpi=80, facecolor='w', edgecolor='k')
    plt.scatter(negDataL['Frame']*0.2*0.001,negDataL[iid],s=.25,label=iid,color='black')
    plt.ylabel(r"Angle ($\degree$) ",color="black",fontsize=25)
    plt.legend(loc='lower left',fontsize=25)
    plt.xlabel(r'Time($\mu$s)',fontsize=25)
    plt.tick_params(axis='x', which='major', labelsize=25)
    plt.tick_params(axis='y', which='major', labelsize=25)
    plt.ylim(0,180)
    plt.xlim(0,2)
    plt.yticks(np.arange(0,200,45))
    plt.show()
```
