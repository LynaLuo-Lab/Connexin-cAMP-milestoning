```python
dataDir='.' #set this to the appropriate directory name
dataFiles=np.sort(os.listdir(dataDir))
```

```python
datFiles=[datFile for datFile in dataFiles if 'NTH' in datFile]
txtFiles=[datFile for datFile in datFiles if 'txt' in datFile]
print txtFiles
parmData=['%s'%restFile for restFile in txtFiles ]

import matplotlib
matplotlib.rc('axes',edgecolor='k')
from matplotlib.pyplot import figure
print "The angle between Principal axis of NTH (1-11) and z-vector "

for restFile in parmData:
    print ('loading fluctuation data from %s'%restFile)
    figure(num=None, figsize=(12, 6), dpi=80, facecolor='w', edgecolor='k')
    negData=pd.read_table(restFile,header=None,delim_whitespace=True)
    negData.columns=['Frame','NTH']
    plt.scatter(negData['Frame']*0.2*0.001,negData['NTH'],s=.25,color='black')
    plt.xlabel(r"Time ($\mu$s)",fontsize=22)
    plt.ylabel(r"Angle (degree)",fontsize=22)
    plt.tick_params(axis='x', which='major', labelsize=28)
    plt.tick_params(axis='y', which='major', labelsize=28)
    plt.yticks(np.arange(0,200,45))
    plt.ylim(0,180)
    plt.xlim(0,2)
    plt.show()
```
