```python
import pytraj as pt
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib
from matplotlib import pyplot as plt
import gc
import mdtraj as md
import pprint,pickle
```

```python
pk1=open('pos200mV_dipoZ.pk1','rb')
dataZ=pickle.load(pk1)
#pprint.pprint(dataZ)
pk1.close()

pk2=open('pos200mV_dipoAngle.pk1','rb')
dataAngle=pickle.load(pk2)
#pprint.pprint(dataAngle)
pk2.close()

lig_residList=[1363,1369,1370,1373,1374,1377,1380,1383]
ligList=['%g'%ligID for ligID in lig_residList]


plt.rcParams['axes.facecolor']='white'
matplotlib.rc('axes',edgecolor='k')
fig,ax1 = plt.subplots(figsize=(10,8))
ax1.grid(False)

for ligI in ligList:
    ax1.scatter(dataZ[ligI],dataAngle[ligI],s=.25)
    ax1.tick_params(axis='x', colors='black')
    ax1.tick_params(axis='y', colors='black')
    ax1.set_ylabel(r"Dipole $\theta$ (degree)",color="black",fontsize=28)
    ax1.set_xlabel(r"Z-axis ($\AA$)",color="black",fontsize=28)
    ax1.tick_params(axis='x', which='major', labelsize=28)
    ax1.tick_params(axis='y', which='major', labelsize=28)
    ax1.set_xlim(-80,80)
    ax1.set_ylim(0,180)
    ax1.set_title("+200mV",fontsize=28)
    ax1.set_xticks(np.arange(-80,100,20))
    ax1.set_yticks(np.arange(0,200,20))
plt.show()
```
